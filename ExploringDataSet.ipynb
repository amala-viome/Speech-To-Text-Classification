```python
import os
import matplotlib.pyplot as plt
```

```python
labels=os.listdir("test")
```

# Labels of Speech Data

```python
print(labels)
print("Total labels: ",len(labels))
```

```python
test_data={}
train_data={}
val_data={}
total_test=0
total_val=0
total_train=0
for label in labels:
    val_data[label]=len(os.listdir("valid/"+label))
    total_val=total_val+val_data[label]
    test_data[label]=len(os.listdir("test/"+label))
    total_test=total_test+test_data[label]
    train_data[label]=len(os.listdir("train/"+label))
    total_train=total_train+train_data[label]
```

# Size of Train, Validation and Test Data

```python
print("Train Data Size: ",total_train)
print("Validation Data Size: ",total_val)
print("Test Data Size: ",total_test)
```

```python
plt.bar(list(train_data.keys()), train_data.values(), color='r')
plt.xlabel("Labels")
plt.ylabel("No. of Audio Files")
plt.title("Train Data Graph")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(list(val_data.keys()), val_data.values(), color='g')
plt.xlabel("Labels")
plt.ylabel("No. of Audio Files")
plt.title("Validation Data Graph")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(list(test_data.keys()), test_data.values(), color='b')
plt.xticks(rotation=90)
plt.title("Test Data Graph")
plt.xlabel("Labels")
plt.ylabel("No. of Audio Files")
plt.show()
```
